```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
```

Captura3.PNG

```python
N = 400
dy = 1/N
y = np.linspace(0, 1, N+1)
```

```python
def mL2V(y):
    return 1000*(y-1/2)**2
```

```python
plt.figure(figsize=(10,5))
plt.plot(y, mL2V(y), lw=2)
plt.title('Potential', fontsize=30)
plt.ylabel('$V/V_0$', fontsize=25)
plt.xlabel('$x/L$', fontsize=25)
plt.grid()
```

```python
d = 1/dy**2 + mL2V(y)[1:-1]
e = -1/(2*dy**2) * np.ones(len(d)-1)
```

$$ M.\vec{\psi} = \alpha . \vec{\psi} \hspace{2cm}, \  \alpha = m.L^2 .E $$

```python
w, v = eigh_tridiagonal(d, e) # nos devuelve los autovalores y autovectores
```

```python
plt.figure(figsize=(10,5))

#graficamos los 4 primeros autoestados 
plt.plot(y[1:-1], v.T[0]**2) # hay que transponer a los autovectores ( en realidad son autofunciones )
plt.plot(y[1:-1], v.T[1]**2)
plt.plot(y[1:-1], v.T[2]**2)
plt.plot(y[1:-1], v.T[3]**2)
plt.title('Eigenstates', fontsize=30)
plt.ylabel('$\psi$', fontsize=25)
plt.xlabel('$x/L$', fontsize=25)
plt.grid()
```

```python
def grafica_proba(ni,nf):
  n_array = list(range(ni,nf))
  ax = plt.axes()
  P=[]
  for i in n_array:
    P.append(v.T[i]**2)
  sum(P)
  plt.plot(y[1:-1],P)
  plt.xlabel("y",fontsize=16)
  plt.ylabel("$ |\psi (x)|^2 $",fontsize=16)
  ax.set(xticks = [])
  plt.grid(alpha=0.9)
```
