## Cordones

```python
db=ds.dropna(subset=['Partido'])
db.value_counts('Partido')


partidos_primer_cordon = [
    'Partido de Avellaneda',
    'Partido de Lanús',
    'Partido de Morón',
    'Partido de Tres de Febrero',
    'Partido de La Matanza','Partido de General San Martín',
    'Partido de Vicente López',
    'Partido de Hurlingham',
      'Partido de Ituzaingó',
    'Partido de San Isidro',"Partido de Lomas de Zamora"
]

segundo_cordon_extendido = [
    'Partido de Berazategui',
    'Partido de Quilmes',
    'Partido de Florencio Varela',
    'Partido de Esteban Echeverría',
    'Partido de Ezeiza',
    'Partido de Moreno',
    'Partido de Merlo',
    'Partido de Malvinas Argentinas',
    'Partido de Hurlingham',

    'Partido de Tigre',
    'Partido de San Fernando',
    'Partido de José C. Paz',
    'Partido de San Miguel',
    'Partido de La Matanza 2',
    'Partido de Almirante Brown',
    'Partido de Islas Tigre',
    #'Partido de Berisso',
    #'Partido de Ensenada',
    #'Partido de Escobar',
    #'Partido de Pilar',
    #'Partido de Marcos Paz',
    #'Partido de Presidente Perón',
    #'Partido de General Rodríguez',
    #'Partido de Luján',
    'Partido de Lomas de Zamora']
partidos_tercer_cordon = [
    'Partido de Pilar','Partido de Escobar','Partido de General Las Heras','Partido de General Rodríguez','Partido de Moreno','Partido de Luján','Partido de Mercedes',
    'Partido de Berisso','Partido de Ensenada','Partido de La Plata','Partido de San Vicente',"Partido de Cañuelas","Partido de Brandsen","Partido de Lobos"
    'Partido de San Andrés de Giles','Partido de Exaltación de la Cruz','Partido de Marcos Paz','Partido de Campana','Partido de Zárate','Partido de Malvinas Argentinas','Partido de Presidente Perón'
]
# Función para realizar la transformación de la matanza
def transformar_partido(df, latitud_limite, longitud_limite):
    for index, row in df.iterrows():
        if row['Partido'] == 'Partido de La Matanza' and row['Latitud'] < latitud_limite and row['Longitud'] < longitud_limite:
            df.at[index, 'Partido'] = 'Partido de La Matanza 2'
    return df
db=transformar_partido(db, -34.70, -58.6)

def f(x):
  for i in partidos_primer_cordon:
    if x==i:
      return "Primer cordon"

  for i in segundo_cordon_extendido:
    if x==i:
      return "Segundo cordon"

  for i in partidos_tercer_cordon:
    if x==i:
      return "Tercer cordon"
  else:
    return "Sin cordon"

db['Cordon']=db['Partido'].apply(lambda x: f(x))

### LIMITES DE CORDONES

gdf_partidos["cordon"]=gdf_partidos['fna'].apply(lambda x: f(x))
gdf_partidos.value_counts('cordon')


sin_c=gdf_partidos[gdf_partidos['cordon']=='Sin cordon']
c1=gdf_partidos[gdf_partidos['cordon']=='Primer cordon']
c2=gdf_partidos[gdf_partidos['cordon']=='Segundo cordon']
c3=gdf_partidos[gdf_partidos['cordon']=='Tercer cordon']

###

import geopandas as gpd
import matplotlib.pyplot as plt
import osmnx as ox

# Obtener el shapefile de Argentina desde el dataset de GeoPandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Filtrar solo Argentina
argentina = world[world.name == 'Argentina']
# Usar osmnx para obtener la geometría de Buenos Aires
buenos_aires = ox.geocode_to_gdf('Buenos Aires, Argentina')
# Graficar
fig, ax = plt.subplots(figsize=(15, 11))

# Graficar Argentina
#argentina.plot(ax=ax, color='white', edgecolor='black')

# Graficar Buenos Aires
import geopandas as gpd
dg = gpd.GeoDataFrame(db, geometry=gpd.points_from_xy(db.Longitud, db.Latitud))
d1=dg.loc[dg['Cordon']=='Primer cordon']
d2=dg.loc[dg['Cordon']=='Segundo cordon']
d3=dg.loc[dg['Cordon']=='Tercer cordon']
d4=dg.loc[dg['Cordon']=='Sin cordon']

#dg.plot(color='white', edgecolor='black')
#sin_c.plot(ax=ax,color="darkred") #grafico las que no tienen cordon

buenos_aires.plot(ax=ax,edgecolor='white',label="CABA")
gdf_partidos.plot(ax=ax,color="lightgrey")
c1.plot(ax=ax,color='lightgrey',edgecolor="coral",linewidth=1.5)
c2.plot(ax=ax,color='lightgrey',edgecolor="lightskyblue",linewidth=1.5)
c3.plot(ax=ax,color='lightgrey',edgecolor="green",linewidth=1.5)

#gdf_partidos.plot(ax=ax, edgecolor='black',color="lightgrey")
d1.plot(ax=ax,color='red',linewidth=1,label="Primer cordon",markersize=6)

d2.plot(ax=ax,color='navy',linewidth=1,label="Segundo cordon",markersize=9)
d3.plot(ax=ax,color='lime',linewidth=1,label="Tercer cordon",markersize=9)
d4.plot(ax=ax,color='fuchsia',linewidth=1,label="Sin cordon",markersize=9)
plt.xlabel("Longitud",fontsize=18)
plt.ylabel("Latitud",fontsize=18)

plt.xlim(-59.5,-57.9)
plt.ylim(-35.3,-34.3)
plt.xticks(fontsize=16)
plt.yticks(fontsize=14)
# x>-58.6 y y>-34.73
# Mostrar el gráfico
plt.grid(alpha=0.5)
plt.legend(fontsize=16)
plt.show()
```
