$$ \Psi(x) = \sum_{n=1} ^ \infty  \ \psi_n (x) =  \sum_{n=1} ^ \infty \frac{1}{\sqrt{2a}} . [ e^{i.k_n .x }.(-1)^{n+1} +e^{-i.k_n .x }] , \ \  \\ k_n = \frac{n \pi }{a} $$

Definimos parametros

````python
a=np.pi
x=np.linspace(-a/2,a/2,200) # dominio
````

Grafica:

 $$ \sum_{n=n_i} ^{n_f}  \ |\psi_n (x)|^2$$

````python
def grafica_proba(ni,nf):
  n_array = list(range(ni,nf))
  prob=[]
  print("Graficamos la sumatoria de probabilidad para n=",list(range(ni,nf)))
  ax = plt.axes()
  plt.text(a/2-0.20,-0.15,'$a/2$',fontsize=18)#texto 
  plt.text(-a/2 -0.25,-0.15,'$ - a/2$',fontsize=18)#texto 
  for n in n_array:
    k=(n * np.pi) / a
    prob.append((abs((1/np.sqrt(2*a)) *(np.exp(k*x*1J) + np.exp(-k*x*1J)* (-1)**(n+1))))**2)
  plt.plot(x,(sum(prob)/np.max(sum(prob))))
  plt.xlabel("x",fontsize=16)
  plt.ylabel("$ |\psi (x)|^2 $",fontsize=16)
  ax.set(xticks = [])
  plt.grid(alpha=0.9)
````

````python
grafica_proba(1,6)
````
